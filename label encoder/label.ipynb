```python
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_data = pd.read_csv("data.csv")
```

```python
cancer_data.head()
```

```python
#finding the counts of the different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder func
label_encoder = LabelEncoder()
```

```python
label= label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
label
```

```python
#appending the labels to thr DF
cancer_data['target'] = label
```

```python
cancer_data['target'].head()
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
